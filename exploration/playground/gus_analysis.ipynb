```python
import pandas as pd
import json
import requests

# Przeciętne miesięczne wynagrodzenia brutto w gospodarce narodowej według województw
url = "https://bdl.stat.gov.pl/api/v1/data/by-variable/64428"
params = {
    'unit-level': 2,  # Voivodships
    'page-size': 100,   # Fetch maximum items per page
    'format': 'json'
}
response = requests.get(url, params=params)
response.raise_for_status()
data = response.json()
```

```python
data
```

```python
print(json.dumps(data, indent=4))
raw_results = data.get('results', [])
print(raw_results)
```

```python
#version of normalizing using pd.json_normalize <-- Better Standard 
df = pd.json_normalize(
        data=raw_results, 
        record_path='values', 
        meta=['id', 'name']
    )
df.describe
```

```python
flattened_data = []

for item in raw_results:
    i_id = item.get('id')
    i_name = item.get('name')
    
    for value in item.get('values', []):
        v_year = value.get('year')
        v_value = value.get('val')
        print(f"ID: {i_id}, Name: {i_name}, Year: {v_year}, Value: {v_value}")
        flattened_data.append({
            'id': i_id,
            'name': i_name,
            'year': v_year,
            'value': v_value
        })
df = pd.DataFrame(flattened_data)
df


            
```
